# Example

```python
vacancy = "We looking for Python ML Engeneer, who can write ML Learing baselines. Bachelon degree required and prostate massage experience required"

resume1 = "Hello, my name is John Doe and I am experienced prostate massage specialist"
resume2 = "I am Good Front End developer, writting on HTML, CSS and PHP"
```

```python
feature_extractor = FeatureExtractor(revision='f2df1d0b06bce3ac1a7ecf7d9408737efb416118') # take other model versions from config.py
sentences_comparator = SortingModel()
```

```python
v, vacancy_embeddings = feature_extractor(vacancy, sentences_comparator)
r1, resume1_embeddings = feature_extractor(resume1, sentences_comparator)
r2, resume2_embeddings = feature_extractor(resume2, sentences_comparator)
```

```python
scores = sentences_comparator.compare_embeddings([vacancy_embeddings, resume1_embeddings]), sentences_comparator.compare_embeddings([vacancy_embeddings, resume2_embeddings])
```

```python
import numpy as np

print(f"""
  {vacancy}

  Matchs more:
  {[resume1, resume2][np.argmax(scores)]}

  Matchs less:
  {[resume1, resume2][~np.argmax(scores)]}
""")
```
